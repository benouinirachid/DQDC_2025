# Outlier Detection

# 2 - Statistical Methods

```python
# prompt: Using data: can you use Z-Score to detect outliers?

import numpy as np

# Calculate the Z-scores
z_scores = np.abs((data - np.mean(data)) / np.std(data))

# Define a threshold for outlier detection (e.g., Z-score > 3)
threshold = 2

# Identify outliers based on the threshold
outliers_zscore = data[z_scores > threshold]

print("Outliers detected using Z-score:")
print(outliers_zscore)

# You can also visualize the outliers along with the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data, bins=25, alpha=0.5, color="green", label='Data')
plt.scatter(outliers_zscore, np.zeros_like(outliers_zscore), color='red', label='Outliers (Z-score)')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data with Outliers (Z-score)')
plt.legend()
plt.show()
```

```python
# prompt: Using data: can you use IQR Method to detect outliers? can you plot the results?

import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is defined as in the provided code
# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers_iqr = data[(data < lower_bound) | (data > upper_bound)]

print("Outliers detected using IQR method:")
print(outliers_iqr)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.hist(data, bins=25, alpha=0.5, color="green", label='Data')
plt.scatter(outliers_iqr, np.zeros_like(outliers_iqr), color='red', label='Outliers (IQR)')
plt.axvline(lower_bound, color='blue', linestyle='dashed', linewidth=2, label='Lower Bound')
plt.axvline(upper_bound, color='blue', linestyle='dashed', linewidth=2, label='Upper Bound')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data with Outliers (IQR)')
plt.legend()
plt.show()
```

```python
# prompt: Using data: can you use Tukey's Fence to detect outlier ? and plot them.

import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is defined as in the provided code (from the previous response)
# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
K = 2
# Define the outlier bounds using Tukey's fences
lower_bound = Q1 - K * IQR
upper_bound = Q3 + K * IQR

# Identify outliers
outliers_tukey = data[(data < lower_bound) | (data > upper_bound)]

print("Outliers detected using Tukey's Fences:")
print(outliers_tukey)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.hist(data, bins=25, alpha=0.5, color="green", label='Data')
plt.scatter(outliers_tukey, np.zeros_like(outliers_tukey), color='red', label='Outliers (Tukey)')
plt.axvline(lower_bound, color='blue', linestyle='dashed', linewidth=2, label='Lower Bound')
plt.axvline(upper_bound, color='blue', linestyle='dashed', linewidth=2, label='Upper Bound')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data with Outliers (Tukey\'s Fences)')
plt.legend()
plt.show()
```
